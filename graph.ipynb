```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vis=pd.read_excel('datacleaned.xlsx')
```

```python
plt.figure(figsize = (9,9)) 
sns.set_style("white")
sns.set_context('paper')
sns.barplot(x='Working Years',y='Perform Evaluation',data=vis,color='#a6cee3',ci=None)
plt.title('Work More Perform More!',fontweight='bold',fontsize=15)
plt.xlabel('Working Years',fontsize=12,fontweight='bold')
plt.ylabel('Perform Evalution',fontsize=12,fontweight='bold')
```

```python
vis.groupby(["Gender", "Marital Status"]).size().unstack(1).plot(kind="barh", stacked=True,color=['#fb9a99','#a6cee3'],figsize=(9,9))
plt.title('Married or Not!',fontweight='bold',fontsize=15)
plt.xlabel('Gender',fontsize=12,fontweight='bold')
plt.ylabel('Marital Status',fontsize=12,fontweight='bold')
sns.set_style("white")
```
